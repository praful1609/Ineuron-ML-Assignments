Ans1. Key tasks in machine learning:
   - Data Collection: Gathering relevant data for analysis.
   - Data Pre-processing: Cleaning, transforming, and organizing data for modeling.
   - Feature Engineering: Creating or selecting relevant features from the data.
   - Model Selection: Choosing an appropriate machine learning algorithm.
   - Model Training: Training the selected model on the data.
   - Model Evaluation: Assessing the model's performance using metrics.
   - Hyperparameter Tuning: Optimizing model parameters for better results.
   - Model Deployment: Deploying the trained model for predictions.
   
   Data pre-processing involves tasks like cleaning noisy data, handling missing values, transforming data into suitable formats, and scaling features to make them compatible with machine learning algorithms.

Ans2. Quantitative and Qualitative Data:
   - Quantitative Data: Represents measurable quantities with numeric values. Examples include age, height, temperature.
   - Qualitative Data: Represents categories or labels with non-numeric values. Examples include gender, color, city names. Qualitative data can further be divided into:
     - Nominal Data: Categories with no inherent order, like colors.
     - Ordinal Data: Categories with a meaningful order, like education levels (e.g., high school, college, PhD).

Ans3. Basic Data Collection (Sample):
   - Machine Learning Data Types:
     - Numeric: Age (e.g., 25, 30, 40)
     - Categorical (Nominal): Color (e.g., Red, Blue, Green)
     - Categorical (Ordinal): Education Level (e.g., High School, College, PhD)
     - Text: Product Reviews (e.g., "Great product!")
     - Image: Photographs (pixel values)
     - Time Series: Stock Prices (e.g., historical price data)

Ans4. Causes of Machine Learning Data Issues:
   - Missing Data: Values that are absent or incomplete.
   - Noisy Data: Data with errors or outliers.
   - Imbalanced Data: Unequal distribution of classes.
   - Inconsistent Data: Contradictory or conflicting information.
   - Data Scaling Issues: Variables on different scales.

   Ramifications include inaccurate models, biased predictions, and decreased model performance.

Ans5. Approaches to Categorical Data Exploration:
   - Frequency Counts: Count the occurrences of each category.
   - Bar Charts: Visualize category frequencies using bars.
   - Pie Charts: Display the distribution of categories in a circular chart.
   - Cross-Tabulations: Analyze relationships between two categorical variables using contingency tables.

Ans6. Impact of Missing Values:
   - Learning may be biased or less effective.
   - Models might not generalize well to new data.

   Solutions include imputation (filling missing values), removing affected records, or using advanced techniques like predictive modeling for imputation.

Ans7. Methods for Dealing with Missing Data:
   - Mean/Median Imputation
   - Mode Imputation
   - Regression Imputation
   - K-Nearest Neighbors Imputation
   - Deletion of Missing Data (rows or columns)
   - Advanced methods like Multiple Imputation

Ans9. i. IQR (Interquartile Range) is a measure of statistical dispersion. It's calculated as Q3 (75th percentile) - Q1 (25th percentile). It's used to assess the spread of data and identify potential outliers.

   ii. Components of a Box Plot: Box (interquartile range), whiskers (minimum and maximum values within 1.5 * IQR), outliers (data points outside the whiskers). The lower whisker surpasses the upper whisker when the lower quartile contains outliers. Box plots can help identify outliers visually.

Ans10. Brief notes:
    - Data collected at regular intervals: Time series data with fixed time intervals, e.g., daily stock prices.
    - Gap between quartiles: The interquartile range (IQR), representing the middle 50% of data.
    - Use a cross-tab: A cross-tabulation table to analyze relationships between two categorical variables.

1. Comparisons:
   - Data with nominal and ordinal values: Nominal data has no order, while ordinal data has a meaningful order.
   - Histogram and box plot: Both visualize data distributions; a histogram shows frequency, while a box plot summarizes quartiles and outliers.
   - Average and median: Both are measures of central tendency, but the average is sensitive to outliers, while the median is robust.
